```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# np.random.seed(8)
x, y = make_blobs(1000,centers=5)
print(x.shape, y.shape)
```

```python
plt.figure(0)
plt.grid(1)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
n = 10*(2*(np.random.random(x.shape[1]))-1)
print(n)
```

```python
k = 5
colors = ['blue','green', 'red', 'orange', 'yellow', 'black']

clusters = {}
for ix in range(k):
    centre = 10*(2*(np.random.random(x.shape[1]))-1)
    points = []
    cluster = {
        'centre': centre,
        'points': points,
        'color' : colors[ix]
    }
    clusters[ix]= cluster
print(clusters)
```

```python
def distance(x1 ,x2):
    return np.sqrt(np.sum((x1-x2)**2))    
```

```python
# def kmeans():
for ix in range(x.shape[0]):
    d= []
    for jx in range(k):
        curr_x = clusters[jx]['centre']
        d.append(distance(curr_x,x[ix]))
    minV = np.argmin(d)
    clusters[minV]['points'].append(x[ix])

for kx in range(k):
    pts = np.array(clusters[kx]['points'])
    clusters[kx]['coords'] = pts
    

plt.figure(0)
plt.grid('True')

for lx in range(k):
    pts = clusters[lx]['coords']
    
    try:
        plt.scatter(pts[:,0],pts[:,1], color = clusters[lx]['color'])
    except:
        pass
    
    center = clusters[lx]['centre']
    plt.scatter(center[0],center[1], color='black',s=100,marker="*")
    
for mx in range(k):
    if clusters[mx]['coords'].shape > 0:
        new_centre = np.mean(clusters[mx]['coords'], axis=0)
    else:
        new_centre = clusters[mx]['centre']

    clusters[mx]['centre'] = new_centre
    
    clusters[mx]['points']= []
    
plt.figure(1)
plt.grid("on")

for kx in range(k):
    pts = clusters[kx]['coords']
    
    try:
        plt.scatter(pts[:,0],pts[:,1], color = clusters[kx]['color'])
    except:
        pass
    
    center = clusters[kx]['centre']
    plt.scatter(center[0],center[1], color='black',s=100,marker="*")
```
